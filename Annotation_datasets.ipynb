```python
# Write the first 510 examples to a text file
n = min(510, len(train_ds))
with open("train_ds.txt", "w") as f:
    for example in train_ds.select(range(n)):
    # for example in train_ds:
        f.write(str(example) + "\n")
```

```python
# importing library
import csv

# Open tsv file and read all rows into a list
with open("data-tokenized.tsv", encoding='latin-1') as tsv_file:
    read_tsv = list(csv.reader(tsv_file, delimiter="\t"))
```

```python
# Write the full data (including headers) to a new TXT file
with open("sample_data.txt", "w") as txt_file:
    for row in read_tsv:
        joined_string = "\t".join(row)
        txt_file.write(joined_string + '\n')
# the code above is to write the full data to txt file, with no modifications
```

```python
in_path = "data-tokenized.tsv"
out_path = "sample_data.txt"

with open(in_path, encoding="latin-1", newline="") as f_in:
    reader = csv.DictReader(f_in, delimiter="\t")
    # if your TSV header is not exactly "ESSAY"/"TEXT", adjust these keys
    rows = []
    for r in reader:
        rows.append([
            r["ESSAY"],   # data point ID output is ESSAY
            r["TEXT"],   # original text from TEXT
        ])

with open(out_path, "w", encoding="utf-8", newline="") as f_out:
    writer = csv.writer(f_out, delimiter="\t")
    writer.writerow(["ID", "TEXT"])
    writer.writerows(rows)
```
