```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("DebateLabKIT/arguments-and-debates", "arguments-pros-and-cons-2010")
```

```python
# Print the first example
print(ds["train"][0])
train_ds = ds["train"]
```

sort the dataset by the length of the argument, then save the first 510 examples to the text file


```python
#sort the dataset by the length of the argument
print(train_ds[0])
```

```python
# Write the first 510 examples to a text file
n = min(510, len(train_ds))
with open("train_ds.txt", "w") as f:
    for example in train_ds.select(range(n)):
    # for example in train_ds:
        f.write(str(example) + "\n")
```
